```python
import re
import json

# Open the file and read its contents
with open('vivaranaDevanagari.txt', 'r') as f:
    vivarana = f.read()

import indic_transliteration

from indic_transliteration import sanscript
from indic_transliteration.sanscript import SchemeMap, SCHEMES, transliterate

import ast

##to get all the available schemes
##indic_transliteration.sanscript.SCHEMES.keys()
    
def transliterateDEVANAGARIAST(text):
    return transliterate(text, sanscript.DEVANAGARI, sanscript.IAST)    
```

```python
vivarana_translit = transliterateDEVANAGARIAST(vivarana)

```

```python
# Open the file in write mode
with open('vivarana_translit.txt', 'w', encoding='utf-8') as f:
    # Write the string to the file
    f.write(vivarana_translit)
```

```python
import re


# Split the text into sections
sections = re.split(r'(.+\|\|.*?\|\|)', vivarana)

print(sections)

# Create a dictionary to hold the sections
section_dict = {}

# Process each section
for section in sections:
    if section.strip() != '':
        # Extract the title, ID, and content
        title, id = re.match(r'(.+?)‌\|\|.*?\|\|', section).groups()
        content = section.split('||', 1)[1].strip()

        # Add the section to the dictionary
        section_dict[id] = {'title': title, 'content': content}

# Now 'section_dict' is a dictionary where each key is a section ID, and each value is another dictionary with the section title and content
for id, section in section_dict.items():
    print(f"ID: {id}\nTitle: {section['title']}\nContent: {section['content']}\n---")
```
